```python
import  mysql.connector as mySQL
```

```python
config = {
    'host': 'localhost',
    'user':'root',
    'password':'root',
    'database':'employees'
}
```

```python
try:
    conn = mySQL.connect(**config)
    cursor = conn.cursor()
    query = "SELECT * FROM emp"
    cursor.execute(query)
    rows= cursor.fetchall()
    for row in rows:
        print(row)
except mySQL.Error as err:
    print(f"Error: {err}")
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'conn' in locals() and conn.is_connected():
        conn.close()
    

```
